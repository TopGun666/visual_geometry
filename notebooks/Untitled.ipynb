```python
import cv2
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns

from src.camera import Camera
from src.utils.cube import generate_cube
```

```python
MATCHER_THREHSOLD = 0.7
```

```python
image_1 = cv2.imread('rec3d_0.png')
image_2 = cv2.imread('rec3d_30.png')
image_3 = cv2.imread('rec3d_50.png')
```

```python
# match features
sift = cv2.xfeatures2d.SIFT_create()
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks = 50)
flann = cv2.FlannBasedMatcher(index_params, search_params)

kp1, des1 = sift.detectAndCompute(image_1, None)
kp2, des2 = sift.detectAndCompute(image_2, None)
kp3, des3 = sift.detectAndCompute(image_3, None)

matches12 = flann.knnMatch(des1, des2, k=2)
matches13 = flann.knnMatch(des1, des3, k=2)
```

```python
# find common query indexes
qidx12 = [m.queryIdx for m,n in list(filter(lambda x: x[0].distance < MATCHER_THREHSOLD*x[1].distance, matches12))]
qidx13 = [m.queryIdx for m,n in list(filter(lambda x: x[0].distance < MATCHER_THREHSOLD*x[1].distance, matches13))]

common_query_indexes = list(set(qidx12).intersection(qidx13))
len(common_query_indexes)
```

```python
# match between 3 images
good_12_matches = []
for m, n in matches12:
    if m.queryIdx in common_query_indexes:
        good_12_matches.append(m)

good_13_matches = []
for m, n in matches13:
    if m.queryIdx in common_query_indexes:
        good_13_matches.append(m)


        
pts1 = np.float64([kp1[m.queryIdx].pt for m in good_12_matches]).reshape(-1,1,2)
pts2 = np.float64([kp2[m.trainIdx].pt for m in good_12_matches]).reshape(-1,1,2)
pts3 = np.float64([kp3[m.trainIdx].pt for m in good_13_matches]).reshape(-1,1,2)
```

```python
pts1.shape, pts2.shape, pts3.shape
```

```python
E, mask = cv2.findEssentialMat(
    pts1, 
    pts2, 
    focal=1.0, 
    pp=(486.2, 265.59), 
    method=cv2.RANSAC, 
    prob=0.999, 
    threshold=1.0
)


points, R, t, mask = cv2.recoverPose(
    E, 
    pts1, 
    pts2, 
    focal=1.0,
    pp=(486.2, 265.59),
    mask=mask
)

K, dist = Camera().K, Camera().distortion 

camera1 = Camera.create(np.eye(3, 3), np.zeros((3, 1)))
camera2 = Camera.create(R, t)

undistorted_points_1 = cv2.undistortPoints(pts1, K, dist, R=np.eye(3, 3), P=camera1.get_projection_matrix())
undistorted_points_2 = cv2.undistortPoints(pts2, K, dist, R=R, P=camera2.get_projection_matrix())

points_3d_homog = cv2.triangulatePoints(
    camera1.get_projection_matrix(),
    camera2.get_projection_matrix(),
    undistorted_points_1,
    undistorted_points_2
)
points_3d = cv2.convertPointsFromHomogeneous(points_3d_homog.T)
```

```python
# get intermediate frame & get common points
image_intermediate = cv2.imread('rec3d_15.png')
kp_intermediate, des_intermediate = sift.detectAndCompute(image_intermediate, None)
matches1intermediate = flann.knnMatch(des1, des_intermediate, k=2)

good_1_intermediate_matches = []
for m, n in matches1intermediate:
    if m.queryIdx in common_query_indexes:
        good_1_intermediate_matches.append(m)
        
pts_intermediate = np.float64([kp_intermediate[m.trainIdx].pt for m in good_1_intermediate_matches]).reshape(-1,1,2)
```

```python
# try to recover intermediate frame
from scipy.optimize import minimize, rosen, rosen_der, basinhopping, fmin, fmin_cg, differential_evolution
from mystic.penalty import quadratic_inequality
from mystic.solvers import diffev2, fmin, fmin_powell, DifferentialEvolutionSolver2
from mystic.monitors import VerboseMonitor
from mystic.termination import VTR
from mystic.strategy import Best1Exp



def reprojection_distance(input, points_3d, points_2d):
    r1, r2, r3, t1, t2, t3 = input[0], input[1], input[2], input[3], input[4], input[5]
    R = cv2.Rodrigues(np.array([r1, r2, r3]))[0]
    t = np.array([[t1], [t2], [t3]])
    camera = Camera()
    projected = cv2.projectPoints(
        points_3d,
        R,
        t,
        camera.K,
        camera.distortion
    )[0]
    vec1 = projected.reshape(projected.shape[0]*2)
    vec2 = points_2d.reshape(points_2d.shape[0]*2)
    res = 0
    for v1, v2 in zip(vec1, vec2):
        res += pow(v1 - v2, 2)
    return res
```

```python
tmp_rod = cv2.Rodrigues(camera1.R)[0]
tmp_tvec = camera1.t
initial = [
    tmp_rod[0],
    tmp_rod[1],
    tmp_rod[2],
    tmp_tvec[0],
    tmp_tvec[1],
    tmp_tvec[2]  
]
initial = [x[0] for x in initial]

upbound = [x*1.5 for x in initial]
lbound = [x*0.5 for x in initial]

bnds = [(x, y) for x, y in zip(upbound, lbound)]

mon = VerboseMonitor(100)
result = diffev2(
    reprojection_distance, 
    x0=bnds, 
    args=(points_3d, pts_intermediate),
    npop=10, 
    gtol=200,
    disp=False, 
    full_output=True, 
    itermon=mon, 
    maxiter=100
)

# mon = VerboseMonitor(5)
# ndim = 6
# npop = 10*ndim
# solver = DifferentialEvolutionSolver2(ndim, npop)
# solver.SetRandomInitialPoints(min=[-100]*ndim, max=[100]*ndim)
# solver.SetGenerationMonitor(mon)
# solver.enable_signal_handler()
# solver.Solve(reprojection_distance, ExtraArgs=(points_3d, pts_intermediate), termination=VTR(0.01), strategy=Best1Exp, \
#                  CrossProbability=1.0, ScalingFactor=0.9)
# solution = solver.Solution()



res = result[0]
r1, r2, r3, t1, t2, t3 = res[0], res[1], res[2], res[3], res[4], res[5]
R_intermediate = cv2.Rodrigues(np.array([r1, r2, r3]))[0]
t_intermediate = np.array([[t1], [t2], [t3]])

camera_intermediate = Camera.create(R_intermediate, t_intermediate)
```

```python
from keras.layers import Input, Dense, Dropout
from keras.models import Model
```

```python
# this is our input placeholder
input_img = Input(shape=(3,))
encoded = Dense(150, activation='relu')(input_img)
encoded = Dropout(0.3)(encoded)
encoded = Dense(64, activation='relu')(encoded)
encoded = Dropout(0.3)(encoded)
encoded = Dense(32, activation='relu')(encoded)

decoded = Dense(32, activation='relu')(encoded)
decoded = Dropout(0.3)(decoded)
decoded = Dense(150, activation='relu')(decoded)
decoded = Dropout(0.3)(decoded)
decoded = Dense(2, activation='relu')(decoded)
autoencoder = Model(input_img, decoded)

autoencoder.compile(optimizer='adagrad', loss='mean_squared_error')
```

```python
from sklearn.model_selection import train_test_split
    
X = points_3d.reshape((points_3d.shape[0], 3))
y = pts_intermediate.reshape((pts_intermediate.shape[0], 2))
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
autoencoder.fit(x_train, y_train,
                epochs=1,
                batch_size=20,
                shuffle=True,
                validation_data=(x_test, y_test)
               )
```
